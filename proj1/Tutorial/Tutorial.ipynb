```python
# É necessário importar os módulos para poder utilisá-los.
# Geralmente são usados aliases (apelidos) mais curtos para simplificar o código
# ex: np ao invés de numpy

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Interpolando strings
# Para colocar o valor de uma variável dentro de uma string, é utilizada a sintaxe de f-string
# As f-strings tem o prefixo 'f' e as variáveis são coilocadas entre chaves {}
var_exemplo = 10
print(f'Isso é uma f-string. O valor de var_exemplo é {var_exemplo}')
```

```python
# Definindo um vetor
v = np.array([0, 1, 2])
print(v)
print(type(v)) # v é uma instância da classe numpy.ndarray

# Calculando algumas propriedades:

# Usando métodos do objeto array
print(v.mean())
print(v.sum())

# Usando funções do módulo numpy
print(np.mean(v))
print(np.sum(v))
```

```python
# Para acessar uma entrada do vetor (a primeira tem o índice 0)
z = np.array([10, 20, 30])
print(z)
print(z[0]) # resultado: 10
# Para modificar um vetor:
z[0] = 42
print(z)
```

```python
# Definindo matrizes
A = np.array([[1, 2],[3, 4]])
B = np.array([[0, 1],[1, 0]])
C = A + B # soma de matrizes
D = A @ B # multiplicação de matrizes

print(f'A = \n {A}')
print(f'B = \n {B}')
print(f'C = A + B = \n {C}')
print(f'D = AB = \n {D}')
```

```python
A = np.array([[1, 2, 3], [10, 20, 30]])
print(A)
# Acessando a primeira linha, segunda coluna:
print(A[0,1]) # Acessando diretamente com um par de índices
print(A[0][1]) # Acessando a linha A[0] depois a segunda coluna
```

```python
v = np.array([1, 2, 3, 4, 5, 6, 7])


print(v[:3]) # Acessando as 3 primeiras entradas
print(v[3:]) # Acessando a partir da 4a entrada
print(v[::2]) # acessando entrada sim, entrada não
print(v[::3]) # acessando entrada sim, 2 entradas não
```

```python
# Produto escalar
u = np.array([1, 2, 3])
v = np.array([1,  10, 100])
x = u.dot(v)
print(x)

# matriz identidade, zeros, uns
print(np.eye(5)) # matriz I 5x5
print(np.ones([2, 3])) # matriz de 1s, 2x3
print(np.zeros([3, 2])) # matriz de 0s, 3x2
print(np.ones(7)) # vetor com 7 entradas, todas =1
```

```python
# Definindo um vetor de 100 entradas igualmente espaçadas entre 0 e 2*pi (incluindo limite superior)
t = np.linspace(0, 2*np.pi, 100)

# Algumas funções podem ser aplicadas em cada uma das entradas, produzindo um novo vetor:
z1 = np.sin(t)
z2 = 0.5 * np.sin(3*t)
# Plotando o resultado
plt.plot(t, z1, color='blue')
plt.plot(t, z2, color='red')
plt.plot(t, z1+z2, color = 'gray')
```

```python
# Para carregar arquivos `.mat` usa-se a função loadmat do módulo scipy.io

variaveis = sp.io.loadmat('data_projeto1.mat')

print(variaveis) # variaveis é um dicionário
sinal = variaveis['sinal']
print(sinal) # sinal é um array
print(sinal) # sua dimensão é 1x1001
sinal = sinal.ravel() # para 'desenrolar' a primeira dimensão e obter um vetor
print(sinal) 
print(sinal.shape) # 1001
```

```python
# para manipular imagens, usamos a biblioteca PILLOW
# consultar o tutorial em https://pillow.readthedocs.io/en/stable/handbook/tutorial.html
from PIL import Image
im = Image.open("lena.bmp")
im # mostra a imagem
```

```python
A = np.array(im, dtype='float') # Retorna um array de float
A = A + 10 # soma 2 em todos os pixels
A = np.array(np.round(A), dtype='uint8')
new_im = Image.fromarray(A)
new_im.save('new_lena.bmp') # salva
new_im # para mostrar
```
